```python
import pandas as pd
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
import seaborn as sns
import seaborn as sb
```

```python
df=pd.read_csv('Weather Dataset.csv')
df.shape
```

```python
pd.isnull(df).sum().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df=df.drop(index=3247)
df
```

```python
df=df.drop(index=236)
```

```python
plt.plot(df['Formatted Date'],df['Temperature (C)'])
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Temp vs Time')
plt.savefig('anything.png')
plt.show()
```

```python
plt.hist(df['Temperature (C)'],bins=30,edgecolor='black')
plt.xlabel('Temperature (C)')
plt.ylabel('Number of time')
```

```python
df['Humidity']
```

```python
plt.scatter(df['Temperature (C)'],df['Humidity'],s=100,edgecolor='black',linewidth=0.55)
plt.xlabel('Temperature (C)')
plt.ylabel('Humidity')
plt.tight_layout()
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number']).corr()
corr_matrix = numeric_df.corr()
sb.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.show()
```
